# My First Eye - in Python ;)

## Getting Started

To keep things simple with the notation, we use `pylab`. This may not be the most *Pythonic* style, but for getting started, it keeps things as simple as possible.

To generate a linear vector, you use `arange`. Note that when you type `arange(` in ipyton notebook and wait for a second, help pops up. If you want more help, type `aranger?` and execute the cell with CTRL+ENTER. You can close the help by clicking on the bar on top of the help window.

In contrast to many other languages, *power* is indicated by `**` and not by `^`.

Also note that Python is completely object oriented. Here for example, `X.shape` tells you the shape of object `X`.

At the end of the first block, we can alreadyshow  our first image - although is is still quite boring: only black.

```python
%pylab inline

# generate a grid
myrange = arange(-250,250)
X,Y = meshgrid(myrange, myrange)
Z = zeros(X.shape)
imshow(Z)
```

## "Iris and Pupil"

$R = \sqrt{x^2 + y^2}$

Note that you can use boolean expressions to address areas with specific properties in the image.

`gray()` sets the colormap to "gray". Type `help(colormaps)` and CTRL+ENTER for more information.

```python
# Iris & pupil
R = sqrt(X**2+Y**2)
Z[R>50]=150
Z[R>100] = 255
gray()
imshow(Z)
```

## Eyelids

The only tricky thing here: I use "int32" to convert the result to integer (we can only set individual points), and use "250+..." to shift things into th middle of our plot.

Here we use for the first time the Python-indentation. Many beginners hate it, I have come to just love it (it keeps things simple and organized at the same time).

```python
# Eyelids
polyPars = polyfit([-250, 0, 250], [0, 100, 0], 2)
x = arange(-250,250)
y = int32(polyval(polyPars, x))

lidThickness = 6
for ii in range(lidThickness):
    Z[ 250 + y+ii, 250 + x] = 1
    Z[ 250 - y-ii, 250 + x] = 1
    
imshow(Z)
```

```python
# Add a few beauty spots
Z[20:40, 20:40] = 0
Z[260:280, 240:250] = 255
imshow(Z)
```

## Saving a figure

We can save the figure to PNG, JPEG, and other formats.

For printing information, I find the notation with `[string].format(content)` the most intuitive to use. If you want, you can add C-type formatting information to the `{0}`.

```python
# Save the figure
print('Z has the shape {0}'.format(Z.shape))
outFile = 'test.png'
imsave(outFile, Z)
```

## Loading a color image

```python
# Load the data again
data = imread(outFile)
print('"data" has the shape {0}'.format(data.shape))
```

## The power of OpenCV

To compare the different approaches, we have to load the individual functions, and do a bit of preparation.

```python
from scipy.ndimage.morphology import binary_closing
from skimage.morphology import closing, square
import cv2

def close(img, strEl):
    ''' Close an image, with a given structuring element. '''
    img_dilated = cv2.dilate(img, strEl)
    img_closed = cv2.erode(img_dilated, strEl)
    return img_closed

bw = data[:,:,0]>0.5
bw_uint8 = uint8(bw*255)
imshow(bw_uint8)

strel = square(20)
```

Now we are at the point where we can compare the three modules. To do so, we use ipython's `%timeit`

```python
print 'Scikit Image:'
%timeit out = closing(bw, strel)

print 'Scipy:'
%timeit out = binary_closing(bw, strel)

print 'OpenCV:'
%timeit out = close(bw_uint8, strel)
```

```python
out = close(bw_uint8, strel)
imshow(out)
```
